```python
# Histogram of housing prices
def houseprice(data, bins=20, color='skyblue', edgecolor='black'):
    plt.hist(data, bins=bins, color=color, edgecolor=edgecolor)
    plt.title('Distribution of Housing Prices')
    plt.xlabel('Price')
    plt.ylabel('Frequency')
    plt.show()

houseprice(hp_df['price'])
```

```python
def houseprice(data, title="Distribution of Housing Prices",
                            xlabel="Price", ylabel="Frequency", figsize=(8, 4)):
    data["price"].plot(kind="line", figsize=figsize, title=title)
    plt.gca().spines[["top", "right"]].set_visible(False)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

houseprice(hp_df)
```

```python
def mainroadPrice(data, x_var, y_var, inner='box', palette='Dark2', title="", figsize=(12, 6)):

  plt.figure(figsize=figsize)
  sns.violinplot(data, x=x_var, y=y_var, inner=inner, palette=palette)
  sns.despine(top=True, right=True, bottom=True, left=True)
  plt.title(title)
  plt.show()

mainroadPrice(hp_df, x_var='price', y_var='mainroad')
```

```python
# Scatter plot of square footage vs. price
def footage_vs_price(data, color='green', alpha=0.5):

    # Scatter plot of 'area' vs. 'price'
    plt.scatter(data['area'], data['price'], color=color, alpha=alpha)
    # Title of the plot
    plt.title('Square Footage vs. Housing Price')
    # X-axis label
    plt.xlabel('Square Footage')
    # Y-axis label
    plt.ylabel('Price')
    # Display the plot
    plt.show()

footage_vs_price(hp_df)
```

```python
# Heatmap of correlation matrix
def correlationmatrix(correlation_matrix, figsize=(10, 8), cmap='coolwarm', fmt=".2f"):

    # Set the figure size
    plt.figure(figsize=figsize)
    # Plot the heatmap with annotations
    sns.heatmap(correlation_matrix, annot=True, cmap=cmap, fmt=fmt)
    # Title of the plot
    plt.title('Correlation Matrix')
    # Display the plot
    plt.show()

correlationmatrix(correlation_matrix)
```
